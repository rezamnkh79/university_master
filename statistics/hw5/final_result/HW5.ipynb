```python
# Q5 part b
import matplotlib.pyplot as plt
import numpy as np

p_values = [0.361, 0.387, 0.005, 0.009, 0.022, 0.051, 0.101, 0.019]

sorted_p_values = sorted(p_values)
ranks = np.arange(1, len(sorted_p_values) + 1)

alpha = 0.05
critical_values = (ranks / len(sorted_p_values)) * alpha

plt.figure(figsize=(8, 6))
plt.plot(ranks, sorted_p_values, label="Sorted p-values", marker='o', color='blue', linestyle='-', markersize=8)

plt.plot(ranks, critical_values, label="Critical values (BH)", color='red', linestyle='--')

plt.axhline(y=alpha, color='green', linestyle='-.', label='Significance threshold (Î±=0.05)')

significant_p_values = [p for i, p in enumerate(sorted_p_values) if p <= critical_values[i]]
significant_ranks = [i+1 for i, p in enumerate(sorted_p_values) if p <= critical_values[i]]
plt.scatter(significant_ranks, significant_p_values, color='green', zorder=5, label="Significant p-values")
plt.xlabel('Rank of p-value')
plt.ylabel('p-value')
plt.title('BH Procedure for FDR Control')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Q6 part a
import pandas as pd
from scipy import stats

lemon_scent = [11, 10, 12]
floral_scent = [11, 14, 11]
fried_food_scent = [5, 5, 7]
no_scent = [8, 7, 6]

data = pd.DataFrame({
    'Lemon Scent': lemon_scent,
    'Floral Scent': floral_scent,
    'Fried Food Scent': fried_food_scent,
    'No Scent': no_scent
})

f_stat, p_value = stats.f_oneway(lemon_scent, floral_scent, fried_food_scent, no_scent)

print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")

if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant difference in the mean number of anagrams solved across different scent treatments.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in the mean number of anagrams solved across different scent treatments.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('final_csv_file1.csv') 
weights = {
    'eccentricity': 0.10, 
    'closeness_centrality': 0.25,  
    'harmonic_closeness': 0.15, 
    'betweenness': 0.25,  
    'num_posts': 0.10,  
    'followers': 0.10 
}

df['new_importance'] = (
    weights['eccentricity'] * df['eccentricity'] +
    weights['closeness_centrality'] * df['closeness_centrality'] +
    weights['harmonic_closeness'] * df['harmonic_closeness'] +
    weights['betweenness'] * df['betweenness'] +
    weights['num_posts'] * df['tweets'] + 
    weights['followers'] * df['followers']  
)
df_sorted = df.sort_values(by='new_importance', ascending=False)
top_10_nodes = df_sorted.head(10)
plt.figure(figsize=(10, 6))
plt.barh(top_10_nodes['label'], top_10_nodes['new_importance'], color='skyblue')
plt.xlabel('Importance')
plt.title('Top 10 Nodes Based on New Importance')
plt.show()
top_10_nodes.to_csv('top_10_nodes_importance.csv', index=False)

```
